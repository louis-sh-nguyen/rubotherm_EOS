# Nomenclature

| $$T$$                  | Temperature $\lbrack K\rbrack$                                                                                                                                 |
|-----------------------|-------------------------------------------------|
| $$P$$                  | Pressure $\lbrack Pa\rbrack$                                                                                                                                   |
| $$S^{EOS}$$            | Solubility of gas in polymer, as calculated by EOS for fully amorphous polymer, defined as ratio of mass of solute to mass of amorphous polymer \[$g/g\rbrack$ |
| $$S$$                  | Overall solubility in semi-crystalline polymer, defined as ratio of mass of solute to mass of semi-crystalline polymer $\lbrack g/g\rbrack\ $                  |
| $$m_{s}$$              | Mass of solute $\lbrack g\rbrack$                                                                                                                              |
| $$m_{p,am}$$           | Mass of amorphous domain of polymer $\lbrack g\rbrack$                                                                                                         |
| $$m_{p,c}$$            | Mass of crystalline domain of polymer $\lbrack g\rbrack$                                                                                                       |
| $$m_{tot}$$            | Total mass (solute + amorphous + crystalline) $\lbrack g\rbrack$                                                                                               |
| $$\omega_{c}$$         | Mass fraction of crystallinity $\lbrack g/g\rbrack$                                                                                                            |
| $${\overline{V}}_{s}$$ | Partial mass volume of solute $\lbrack g\rbrack$                                                                                                               |
| $${\overline{V}}_{p}$$ | Partial mass volume of polymer $\lbrack g\rbrack$                                                                                                              |
| $$V_{am}$$             | Total volume of amorphous domain of polymer $\left\lbrack m^{3} \right\rbrack$                                                                                 |
| $$V_{tot}$$            | Total volume of polymer $\left\lbrack m^{3} \right\rbrack$                                                                                                     |
| $${\overline{V}}_{s}$$ | Partial mass volume of solute $\left\lbrack m^{3}/g \right\rbrack$                                                                                             |
| $${\overline{V}}_{p}$$ | Partial mass volume of polymer $\left\lbrack m^{3}/g \right\rbrack$                                                                                            |
| $$\rho_{p,am}$$        | Denisty of polymer in amorphous domain $\left\lbrack g/m^{3} \right\rbrack$                                                                                    |
| $$\rho_{p,c}$$         | Density of crystalline polymer $\left\lbrack g/m^{3}\  \right\rbrack$                                                                                          |
| $$\rho_{f}$$           | Density of external fluid $\left\lbrack g/m^{3} \right\rbrack$                                                                                                 |
| $$F$$                  | Net force $\lbrack Pa\rbrack$                                                                                                                                  |
| $$F_{w}$$              | Weight $\lbrack Pa\rbrack$                                                                                                                                     |
| $$F_{B}$$              | Buoyancy force $\lbrack Pa\rbrack$                                                                                                                             |
| $$g$$                  | Gravitational acceleration $\left\lbrack m/s^{2} \right\rbrack$                                                                                                |

# Equations 

### Definitions

Definition for crystallinity:

$$\omega_{c} = \frac{m_{p,c}}{m_{p,tot}}$$

Definition for solubility in amorphous polymer, calculated from EOS:

$$S^{EoS}(T,P) = \frac{m_{s}}{m_{p,am}}$$

Definition for overall solubility:

$$S(T,P) = \frac{m_{s}}{m_{p,tot}} = \left( 1 - \ \omega_{c} \right)S^{EoS}(T,P)$$

### Calculate total density $\rho_{tot}$

Total volume in amorphous domain of polymer:

$$V_{am} = m_{s}*{\overline{V}}_{s}\left( T,P,S^{EoS} \right) + m_{p,am}\ {\overline{V}}_{p}\left( T,P,S^{EoS} \right)$$

divide both sides by $m_{p,am}$**:**

$$\frac{1}{{\rho_{p},}_{am}\left( T,P,S^{EoS} \right)} = \left( S^{EoS}{\overline{V}}_{s}\left( T,P,S^{EoS} \right) + {\overline{V}}_{p}\left( T,P,S^{EoS} \right) \right)$$

Total volume of polymer (amorphous + crystalline):

$$V_{tot} = V_{am} + V_{c} = \frac{m_{p,am}}{\rho_{p,am}} + \frac{m_{p,c}}{\rho_{p,c}}$$

divide both sides by $m_{tot}$:

$$\frac{1}{\rho_{tot}} = \frac{m_{p,am}}{\rho_{p,am}m_{tot}} + \frac{m_{p,c}}{\rho_{c}m_{tot}}$$

and because:

$$\frac{m_{p,am}}{m_{tot}} = \frac{m_{p,am}}{m_{p,tot} + m_{s}} = \frac{1 - \omega_{c}}{1 + S}$$

$$\frac{m_{p,c}}{\rho_{c}m_{tot}} = \frac{m_{p,c}}{\rho_{c}m_{p,tot}}\frac{m_{p,tot}}{m_{tot}} = \frac{\omega_{c}}{\rho_{c}(1 + S)}\ $$

$$\frac{1}{\mathbf{\rho}_{\mathbf{tot}}} = \frac{1}{1 + S}\left( \frac{1 - \omega_{c}}{\rho_{p,am}} + \frac{\omega_{c}}{\rho_{c}} \right)$$

Force balance:

$$F = F_{w} - F_{B}$$

combining with

$$F_{w} = m_{tot}*g$$

$$F_{B} = V_{tot}*\rho_{f}(T,P)*g = m_{tot}*\frac{\rho_{f}}{\rho_{tot}}*g$$

gives

$$F = \left( m_{tot} - V_{tot}*\rho_{f} \right)g$$

Then, divide both sides by $m_{p,tot}$:

$$\frac{F}{m_{p,tot}} = (1 + S)\left( 1 - \frac{\rho_{f}}{\rho_{tot}} \right)g = \left( 1 + \left( 1 - \omega_{c} \right)S^{EoS} \right)\ \left( 1 - \frac{\rho_{f}(T,P)}{\rho_{tot}\left( T,P,S^{EoS},\omega_{c},\rho_{c} \right)} \right)g$$

Correction of force balance dependent on calculation of total density
$\rho_{tot}$, which ultimately dependent on evaluation of partial volume
${\overline{V}}_{s}$ and ${\overline{V}}_{p}$.

Evaluation of partial volume ${\overline{V}}_{s}$ and
${\overline{V}}_{p}$:

$${\overline{V}}_{s} = {\overline{V}}_{s}\left( T,P,n_{s},n_{p} \right)$$

$${\overline{V}}_{p} = {\overline{V}}_{p}\left( T,P,n_{s},n_{p} \right)$$

Definition of partial quantity:

$${\overline{V}}_{s} = \left( \frac{\partial V}{\partial n_{s}} \right)_{T,P,n_{p}}$$

$${\overline{V}}_{p} = \left( \frac{\partial V}{\partial n_{p}} \right)_{T,P,n_{s}}$$

**Method 1 (most robust)**: ${\overline{V}}_{s}$ and
${\overline{V}}_{p}$ are calculated at each T, P and solubility $S$.

$${\overline{V}}_{s}\sim\frac{V\left( T,P,n_{s} + dn_{s},n_{p} \right) - V\left( T,P,n_{s},n_{p} \right)}{dn_{s}}$$

$${\overline{V}}_{p}\sim\frac{V\left( T,P,n_{s},n_{p} + dn_{p} \right) - V\left( T,P,n_{s},n_{p} \right)}{dn_{p}}$$

> **Method 2**: ${\overline{V}}_{s}$ and ${\overline{V}}_{p}$ are
> calculated at infinitely dilute condition of solute. This assumes
> ${\overline{V}}_{s}$ and ${\overline{V}}_{p}$ are dependent on T and P
> only, and not on solubility S.

$${\overline{V}}_{s}\sim\frac{V\left( T,P,dn_{s},n_{p} \right) - V\left( T,P,0,n_{p} \right)}{dn_{s}}$$

$${\overline{V}}_{p}\sim\frac{V\left( T,P,0,n_{p} + dn_{p} \right) - V\left( T,P,0,n_{p} \right)}{dn_{p}}$$

> **Method 3 (least robust):** ${\overline{V}}_{s}$ and
> ${\overline{V}}_{p}$ are unchanged in the condensed phase inside the
> polymer and calculated at 1 bar. As a result, ${\overline{V}}_{s}$ and
> ${\overline{V}}_{p}$ are dependent of T only, and not on P and
> solubility $S$.

$${\overline{V}}_{s}\sim\frac{V\left( T,1\ bar,dn_{s},n_{p} \right) - V\left( T,1\ bar,0,n_{p} \right)}{dn_{s}}$$

$${\overline{V}}_{p}\sim\frac{V\left( T,1\ bar,0,n_{p} + dn_{p} \right) - V\left( T,1\ bar,0,n_{p} \right)}{dn_{p}}$$
